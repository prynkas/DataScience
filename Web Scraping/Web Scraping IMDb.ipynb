```python
url="https://www.flipkart.com/hp-15s-ryzen-3-dual-core-3250u-4-gb-1-tb-hdd-windows-10-home-15s-gr0006au-thin-light-laptop/product-reviews/itmdbb2d234797b6?pid=COMFZHZFZFFBJBHH&lid=LSTCOMFZHZFZFFBJBHH7WNMPE&marketplace=FLIPKART"
```

```python
from bs4 import BeautifulSoup as bs #beautifulsoup is a library
```

```python
import requests # requests is a module
```

```python
url="https://www.flipkart.com/hp-15s-ryzen-3-dual-core-3250u-4-gb-1-tb-hdd-windows-10-home-15s-gr0006au-thin-light-laptop/product-reviews/itmdbb2d234797b6?pid=COMFZHZFZFFBJBHH&lid=LSTCOMFZHZFZFFBJBHH7WNMPE&marketplace=FLIPKART"
```

```python
r = requests.get(url)
r
```

```python
htmlcontent = r.content  #getting html content into python code
htmlcontent
```

```python
soup = bs(htmlcontent,"html.parser")
```

```python
print(soup.prettify())
```

# Title of product

```python
title=soup.title
print(title)
```

```python
print(title.string)
```

### Check product price

```python
price=soup.find_all("div",class_="_30jeq3")
price
```

```python
product_price=[]
for i in range(0,len(price)):
    product_price.append(price[i].get_text())
product_price
```

### Get all customers name

```python
names = soup.find_all("p",class_="_2sc7ZR _2V5EHH")
names
```

```python
customer_name=[]
for i in range(0,len(names)):
    customer_name.append(names[i].get_text())
customer_name
```

### Get all customers comments

```python
comment = soup.find_all("p",class_="_2-N8zT")
comment
```

```python
cust_comment=[]
for i in range(0,len(comment)):
    cust_comment.append(comment[i].get_text())
cust_comment
```

```python
star = soup.find_all("div",class_="_3LWZlK _1BLPMq")
star
```

```python
cust_star=[]
for i in range(0,len(star)):
    cust_star.append(star[i].get_text())
cust_star
```

### get all customers review

```python
review=soup.find_all("div",class_="t-ZTKy")
review
```

```python
cust_review=[]
for i in range(0,len(review)):
    cust_review.append(review[i].get_text())
cust_review
```

```python
import pandas as pd
```

```python
df=pd.DataFrame()
df["Customer_Name"] = customer_name
df["comments"] = cust_comment
df["star"] = cust_star
df["Reviews"] = cust_review
df
```
